#capturing nan value with the new feature and variable
it works well if the data are not missing completely at random

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('C:\\Users\\adarsh\\Untitled Folder\\titanic_train.csv', usecols=['age','fare','survived'])
df.head()
```

```python
df['age_nan']=np.where(df['age'].isnull(),1,0)
df.head()
```

```python
df.age.median()  #you can also take as ean
```

```python
df['age'].fillna(df.age.median(),inplace = True)    #inplace is used for making the change permanently
```
